```python
#Importing numpy as its alias
import numpy as np
```

```python
#Creating a numpy array
np_array = np.array([1,2,3,4])
np_array
type(np_array)
```

```python
#Multiplying every number in a array by 2

#In python
numbers = [1,2,3,4,5]
doubled = []

for num in numbers:
    doubled.append(num * 2)
print(doubled)
```

```python
#multiplying a numpy array by 2

np_array = np.array([doubled])
np_doubled = np_array * 2
np_doubled

```
